## 1. What is statistics, and why is it important?

* Statistics is the science of collecting and studying data.
* It helps us understand information in an organized way.
* We use statistics to find patterns and trends in data.
* It helps in decision-making based on facts, not guesses.
* Statistics is used in business, education, medicine, and research.
* It helps reduce uncertainty and improve accuracy.

## 2. What are the two main types of statistics?

* There are **two main types** of statistics.
* They are **Descriptive statistics** and **Inferential statistics**.
* Descriptive statistics summarize data clearly.
* Inferential statistics make predictions using data.
* Both types help in understanding large data sets.
* They are used together in real-life analysis.

## 3. What are descriptive statistics?

* Descriptive statistics describe and summarize data.
* They make large data easy to understand.
* Examples include mean, median, mode, charts, and tables.
* They do not predict future results.
* They only explain what the data shows.
* Used in reports, exams, and surveys.

## 4. What is inferential statistics?

* Inferential statistics help make predictions.
* They use sample data to draw conclusions about a population.
* Examples include hypothesis testing and confidence intervals.
* It helps in decision-making under uncertainty.
* Widely used in research and experiments.
* Results are based on probability.

## 5. What is sampling in statistics?

* Sampling is selecting a small part of a population.
* The sample represents the whole population.
* It saves time, cost, and effort.
* Used when studying the whole population is difficult.
* Good sampling gives accurate results.
* Example: surveying 100 students from a college.

## 6. What are the different types of sampling methods?

* Sampling methods are **random** and **non-random**.
* Random sampling gives equal chance to all.
* Non-random sampling does not give equal chance.
* Examples: simple random, stratified, systematic.
* Non-random examples: convenience, judgment sampling.
* Choice depends on the study type.

## 7. Difference between random and non-random sampling?

* Random sampling is unbiased and fair.
* Every individual has equal chance of selection.
* Non-random sampling depends on choice or convenience.
* Random sampling gives more accurate results.
* Non-random sampling may cause bias.
* Random is preferred in research.

## 8. Qualitative and quantitative data with examples

* Qualitative data describes qualities or categories.
* It is non-numeric data.
* Example: gender, color, type of food.
* Quantitative data is numerical.
* It can be measured or counted.
* Example: height, marks, age.

## 9. Different types of data in statistics

* Data is mainly **qualitative** and **quantitative**.
* Qualitative data is descriptive.
* Quantitative data is numerical.
* Quantitative data can be discrete or continuous.
* Discrete: number of students.
* Continuous: weight, height.

## 10. Nominal, ordinal, interval, and ratio scales

* Nominal data has no order (gender, color).
* Ordinal data has order but no fixed difference (ranks).
* Interval data has equal intervals but no true zero (temperature).
* Ratio data has equal intervals and true zero (height, weight).
* Each level adds more information.
* Used to choose correct statistical methods.

## 11. What is the measure of central tendency?

* It shows the center or average of data.
* It helps summarize data in one value.
* Common measures are mean, median, and mode.
* It represents typical value of data.
* Easy to understand and compare.
* Used in reports and analysis.

## 12. Define mean, median, and mode

* **Mean** is the average of all values.
* Mean = sum of values ÷ total values.
* **Median** is the middle value.
* Used when data is ordered.
* **Mode** is the most frequent value.
* A dataset can have more than one mode.

## 13. Significance of measure of central tendency

* It summarizes large data in one number.
* Makes comparison between datasets easy.
* Helps understand data behavior quickly.
* Used in education, business, and research.
* Reduces complexity of raw data.
* Helps in decision-making.

## 14. What is variance and how is it calculated?

* Variance measures how data spreads from the mean.
* It shows how much values differ.
* Calculate by finding squared differences from mean.
* Then divide by total number of values.
* High variance means data is spread out.
* Low variance means data is close to mean.

## 15. What is standard deviation and why is it important?

* Standard deviation is the square root of variance.
* It shows data spread in same unit as data.
* Easy to understand compared to variance.
* Used in exams, finance, and science.
* Helps measure consistency.
* Lower value means more consistency

## 16. Define range in statistics

* Range is the difference between highest and lowest value.
* Range = Maximum value − Minimum value.
* It shows how wide data is spread.
* Very easy to calculate.
* Used for quick analysis.
* Not affected by all values.

## 17. Difference between variance and standard deviation

* Variance uses squared units.
* Standard deviation uses original units.
* Variance is harder to interpret.
* Standard deviation is easier to understand.
* Both measure data spread.
* Standard deviation is more commonly used.

## 18. What is skewness in a dataset?

* Skewness shows the shape of data distribution.
* It tells if data is symmetrical or not.
* Data can be positively or negatively skewed.
* Helps understand data behavior.
* Used in probability and analysis.
* Important for interpretation.

## 19. Positive and negative skewness meaning

* Positive skew means tail is on the right side.
* Mean is greater than median.
* Negative skew means tail is on the left side.
* Mean is less than median.
* Shows imbalance in data.
* Common in income and exam scores.

## 20. Define and explain kurtosis

* Kurtosis shows how peaked data is.
* It measures sharpness of distribution.
* High kurtosis means sharp peak.
* Low kurtosis means flat distribution.
* Helps understand extreme values.
* Used in data analysis.

## 21. Purpose of covariance

* Covariance shows relationship between two variables.
* It tells if variables move together.
* Positive covariance means both increase together.
* Negative means one increases, other decreases.
* Does not show strength clearly.
* Used in finance and statistics.

## 22. What does correlation measure?

* Correlation measures strength of relationship.
* Values range from -1 to +1.
* +1 means perfect positive relation.
* -1 means perfect negative relation.
* 0 means no relationship.
* Used in prediction and analysis.

## 23. Difference between covariance and correlation

* Covariance shows direction of relationship.
* Correlation shows direction and strength.
* Covariance has no fixed range.
* Correlation is always between -1 and +1.
* Correlation is easier to interpret.
* Both study variable relationships.

## 24. Real-world applications of statistics

* Used in business for profit analysis.

* Helps governments in planning policies.

```python
#Calculate the mean, median, and mode of grouped data5
import pandas as pd

# 1. Setup the Grouped Data
data = {
    'Lower_Limit': [0, 10, 20, 30],
    'Upper_Limit': [10, 20, 30, 40],
    'Frequency': [5, 10, 15, 10]
}
df = pd.DataFrame(data)

# 2. Calculate Mean
# Formula: Sum(f * mid_point) / Sum(f)
df['Mid_Point'] = (df['Lower_Limit'] + df['Upper_Limit']) / 2
df['f_x_m'] = df['Frequency'] * df['Mid_Point']
mean_grouped = df['f_x_m'].sum() / df['Frequency'].sum()

# 3. Calculate Median
# Formula: L + ((N/2 - CF_prev) / f_median) * h
N = df['Frequency'].sum()
df['Cumulative_Freq'] = df['Frequency'].cumsum()
h = df['Upper_Limit'][0] - df['Lower_Limit'][0] # Class width

# Find Median Class (where Cumulative Frequency >= N/2)
median_class = df[df['Cumulative_Freq'] >= N/2].iloc[0]
L_median = median_class['Lower_Limit']
f_median = median_class['Frequency']
# Get CF of the class before the median class
idx = df[df['Cumulative_Freq'] >= N/2].index[0]
CF_prev = df['Cumulative_Freq'][idx-1] if idx > 0 else 0

median_grouped = L_median + (((N/2) - CF_prev) / f_median) * h

# 4. Calculate Mode
# Formula: L + ((f1 - f0) / (2f1 - f0 - f2)) * h
# Find Modal Class (class with highest frequency)
modal_idx = df['Frequency'].idxmax()
modal_class = df.loc[modal_idx]

L_mode = modal_class['Lower_Limit']
f1 = modal_class['Frequency']
f0 = df['Frequency'][modal_idx-1] if modal_idx > 0 else 0
f2 = df['Frequency'][modal_idx+1] if modal_idx < len(df)-1 else 0

mode_grouped = L_mode + ((f1 - f0) / (2*f1 - f0 - f2)) * h

print(f"Grouped Mean: {mean_grouped}")
print(f"Grouped Median: {median_grouped}")
print(f"Grouped Mode: {mode_grouped}")
```

```python
# Simulate data using Python and calculate its central tendency and dispersion.
import numpy as np
import pandas as pd
from scipy import stats

# 1. Simulate Data: 1000 samples with mean=50 and standard deviation=10
np.random.seed(42)  # For reproducibility
simulated_data = np.random.normal(loc=50, scale=10, size=1000)

# 2. Calculate Central Tendency
mean = np.mean(simulated_data)
median = np.median(simulated_data)
mode_result = stats.mode(simulated_data, keepdims=True)
mode = mode_result.count[0]

# 3. Calculate Dispersion
data_range = np.ptp(simulated_data)  # Peak-to-Peak (Max - Min)
variance = np.var(simulated_data)
std_dev = np.std(simulated_data)
iqr = stats.iqr(simulated_data)

# Display Results
print("--- Central Tendency ---")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")

print("\n--- Dispersion ---")
print(f"Range: {data_range:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#  Generate a dataset and implement both variance and standard deviation computations
import numpy as np

# 1. Generate a dataset
# Let's simulate the weights (in kg) of 10 individuals
weights = np.array([65, 70, 72, 68, 80, 55, 92, 67, 75, 71])

# --- Method 1: Manual Computation (Step-by-Step) ---
mean = np.mean(weights)

# Variance = Sum of (x - mean)^2 / N
squared_diff = (weights - mean) ** 2
manual_variance = np.sum(squared_diff) / len(weights)

# Standard Deviation = Square root of Variance
manual_std_dev = np.sqrt(manual_variance)

# --- Method 2: NumPy Computation (Efficient) ---
numpy_variance = np.var(weights)
numpy_std_dev = np.std(weights)

# Display Results
print(f"Dataset: {weights}")
print(f"Mean: {mean:.2f}")
print("-" * 30)
print(f"Manual Variance: {manual_variance:.2f}")
print(f"Manual Std Dev: {manual_std_dev:.2f}")
print("-" * 30)
print(f"NumPy Variance: {numpy_variance:.2f}")
print(f"NumPy Std Dev: {numpy_std_dev:.2f}")
```

```python
#  Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# 1. Generate Datasets
np.random.seed(42)
n = 1000

# Normal Distribution (Reference: Skew ≈ 0, Excess Kurtosis ≈ 0)
data_normal = np.random.normal(0, 1, n)

# Positively Skewed (Tail on the right)
data_skewed = np.random.exponential(1, n)

# Leptokurtic (High Kurtosis - Sharp peak and thick tails)
data_kurtotic = np.random.laplace(0, 1, n)

# 2. Setup Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Normal
sns.histplot(data_normal, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title(f"Normal Distribution\nSkew: {skew(data_normal):.2f}, Kurtosis: {kurtosis(data_normal):.2f}")

# Plot 2: Positive Skew
sns.histplot(data_skewed, kde=True, ax=axes[1], color='salmon')
axes[1].set_title(f"Positively Skewed\nSkew: {skew(data_skewed):.2f}, Kurtosis: {kurtosis(data_skewed):.2f}")

# Plot 3: High Kurtosis
sns.histplot(data_kurtotic, kde=True, ax=axes[2], color='lightgreen')
axes[2].set_title(f"High Kurtosis (Leptokurtic)\nSkew: {skew(data_kurtotic):.2f}, Kurtosis: {kurtosis(data_kurtotic):.2f}")

plt.tight_layout()
plt.show()
```

```python
# Implement the Pearson and Spearman correlation coefficients for a dataset.
import pandas as pd
from scipy.stats import pearsonr, spearmanr

data1 = [1, 2, 3, 4, 5]
data2 = [1, 4, 9, 16, 25] # Non-linear relationship ($y = x^2$)

p_corr, _ = pearsonr(data1, data2)
s_corr, _ = spearmanr(data1, data2)

print(f"Pearson (Linear): {p_corr:.3f}")
print(f"Spearman (Rank): {s_corr:.3f}") # Spearman handles the curve better
```
