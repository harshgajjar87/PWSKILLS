1. What are data structures, and why are they important?

Ans. Data structures are ways of organizing and storing data in a computer so it can be accessed and used efficiently. They are important because they allow for efficient data management, enhanced program performance, and help simplify complex problems, forming the foundation for many advanced computer science topics.

2. Explain the difference between mutable and immutable data types with examples.

Ans. Mutable data types can be changed after they are created (e.g., lists, dictionaries, sets), while immutable data types cannot be altered once defined (e.g., strings, tuples, integers, floats). Any "modification" to an immutable object actually creates a new object in memory, ensuring data integrity.


3. What are the main differences between lists and tuples in Python?

Ans. The main difference is mutability; lists are mutable (changeable), and tuples are immutable (unchangeable). Lists are defined using square brackets [], while tuples use parentheses (). Tuples are generally faster and more memory-efficient than lists for static data.  

4. Describe how dictionaries store data.

Ans. Dictionaries store data in key-value pairs within curly braces {}. They use a hash table implementation internally, where each unique, immutable key is mapped to an index (or "bucket") using a hash function, allowing for very fast data retrieval.


5. Why might you use a set instead of a list in Python?

Ans. You would use a set over a list to store only unique elements or when you need fast membership testing (checking if an item exists in the collection). Sets automatically handle duplicate removal and offer O(1) average time complexity for membership checks, which is much faster than the O(n) complexity of lists.


6. What is a string in Python, and how is it different from a list?

Ans. A string in Python is an ordered, immutable sequence of characters. Unlike lists, which are mutable collections of elements, individual characters within a string cannot be changed after creation; operations appear to modify the string but instead create a new string object.


7. How do tuples ensure data integrity in Python?

Ans.  Tuples ensure data integrity through their immutability. Once a tuple is created, its elements cannot be modified, added, or removed, which prevents accidental data changes and makes them safe for data that should remain constant throughout a program.

8. What is a hash table, and how does it relate to dictionaries in Python?

Answer: A hash table is a data structure that maps keys to values using a hash function for fast insertion, retrieval, and deletion. In Python, dictionaries are the built-in implementation of hash tables.


9. Can lists contain different data types in Python?

Answer: Yes, lists in Python are heterogeneous and can contain different data types, including numbers, strings, or even other lists, tuples, and dictionaries.

10. Explain why strings are immutable in Python.

Answer: Strings are immutable to ensure data integrity and allow them to be used safely as dictionary keys or set elements, as their hash value remains constant throughout their lifetime. This design also provides inherent thread-safety in multi-threaded environments.

11. What advantages do dictionaries offer over lists for certain tasks?

Answer: Dictionaries offer significantly faster lookups and data retrieval based on a key (average O(1) time complexity) compared to lists (average O(n) time complexity). They are ideal for storing data with associated labels (keys) rather than just a sequential order.

12. Describe a scenario where using a tuple would be preferable over a list.

Answer: A tuple would be preferable for storing fixed, unchangeable data like a geographic coordinate (latitude, longitude), an RGB color value (255, 0, 0), or constant configuration settings. Its immutability guarantees the data remains consistent.

13. How do sets handle duplicate values in Python?

Answer: Sets automatically ensure all elements are unique. If you try to add a duplicate value to a set, it will simply be ignored, and only one instance of the value will be stored.

14. How does the "in" keyword work differently for lists and dictionaries?

Answer: The in keyword checks for an element's presence in a list (which is a linear scan, average O(n)) versus checking for a key's presence in a dictionary (which is a hash lookup, average O(1)). The dictionary check is significantly faster, especially for large collections.

15. Can you modify the elements of a tuple? Explain why or why not.

Answer: No, individual elements of a tuple cannot be modified because tuples are an immutable data type. Any attempt to change an element will result in a TypeError.

16. What is a nested dictionary, and give an example of its use case?

Answer: A nested dictionary is a dictionary that contains another dictionary as one or more of its values. A common use case is storing structured, hierarchical data, such as a contact list where the outer key is a name, and the value is another dictionary containing details like 'phone', 'email', and 'address'.

17. Describe the time complexity of accessing elements in a dictionary.

Answer: The average time complexity for accessing elements (using a key) in a Python dictionary is constant time, O(1). This efficiency is due to its underlying hash table implementation.

18. In what situations are lists preferred over dictionaries?

Answer: Lists are preferred when the order of elements matters and needs to be preserved, when you need to access elements by a numerical index, or when the data will be frequently modified with additions or deletions.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?

Answer: As of Python version 3.7, dictionaries maintain insertion order, but they were historically unordered because their elements are stored based on hash values, not sequential indices. This means you cannot rely on index-based access, but it enables the fast, key-based retrieval that defines their performance.

20. Explain the difference between a list and a dictionary in terms of data retrieval.

Answer: Lists use integer indices for data retrieval (e.g., my_list[0]), requiring a linear search through elements if the index is unknown. Dictionaries use unique, hashable keys for retrieval (e.g., my_dict['key']), allowing for near-instant (O(1) average time) access to the associated value.
