1. What is Object-Oriented Programming (OOP)?

OOP is a programming paradigm that structures code around "objects" (data) rather than just functions (logic). It combines data (attributes) and behavior (methods) into single units to model real-world entities.

2. What is a class in OOP?

A class is a blueprint or template for creating objects. It defines the structure (attributes) and behaviors (methods) that the objects created from it will possess.

3. What is an object in OOP?

An object is a specific instance of a class. If the class is the blueprint for a house, the object is the actual house built from that blueprint, containing real data.

4. What is the difference between abstraction and encapsulation?


Abstraction is about hiding complexity (showing only what is necessary), while Encapsulation is about data protection (bundling data and methods and restricting access). Abstraction is the design level; Encapsulation is the implementation level.

5. What are dunder methods in Python?

Dunder (Double UNDERscore) methods, or "Magic Methods" (like __init__, __str__), allow objects to interact with Python's built-in behavior. They let you define how your objects handle operations like addition, printing, or length checking.

6. Explain the concept of inheritance in OOP.

Inheritance allows a child class to acquire the properties and methods of a parent class. This promotes code reusability, as you don't need to rewrite code for shared behaviors.

7. What is polymorphism in OOP?

Polymorphism means "many forms." It allows different classes to be treated as instances of the same general class through the same method name (e.g., both Dog and Cat have a speak() method, but they act differently).

8. How is encapsulation achieved in Python?

It is achieved using underscores. A single underscore (_name) implies "protected" (internal use), while a double underscore (__name) triggers name mangling to make the variable private and harder to access from outside.

9. What is a constructor in Python?


The __init__ method acts as the constructor. It is automatically executed when a new object is created to initialize the object's attributes (state).

10. What are class and static methods in Python?

A Class Method (@classmethod) takes cls as the first argument and affects the entire class. A Static Method (@staticmethod) behaves like a regular function inside a class and doesn't access specific object or class data.

11. What is method overloading in Python?

Python does not support traditional method overloading (same function name, different parameters). However, we can achieve similar behavior using default arguments (def add(a, b=0):) or variable-length arguments (*args).

12. What is method overriding in OOP?

Method overriding happens when a child class provides a specific implementation of a method that is already defined in its parent class. The child's version replaces the parent's version for that object.

13. What is a property decorator in Python?

The @property decorator allows you to define a method that can be accessed like a variable (without parentheses). It is often used to implement "getter" and "setter" logic for Encapsulation.

14. Why is polymorphism important in OOP?

It adds flexibility and makes code extendable. You can write a function that works with a general "Animal" type, and it will automatically work with any new "Animal" subclass you create in the future without changing the code.

15. What is an abstract class in Python?

An abstract class is a blueprint that cannot be instantiated (you can't make an object from it). It contains abstract methods that force child classes to implement specific behaviors.

16. What are the advantages of OOP?

It provides modularity (easier to troubleshoot), reusability (via inheritance), security (via encapsulation), and flexibility (via polymorphism), making large software easier to manage and scale.

17. What is the difference between a class variable and an instance variable?

A Class Variable is shared by all objects of that class (changing it changes it for everyone). An Instance Variable belongs to one specific object (changing it only affects that object).

18. What is multiple inheritance in Python?

Multiple inheritance is when a child class inherits from more than one parent class at the same time. This allows the child to combine features from distinct sources (e.g., a FlyingCar inheriting from both Car and Airplane).

19. Explain the purpose of __str__ and __repr__ methods.

__str__ returns a readable string for end-users (what you see when you print()). __repr__ returns an unambiguous string meant for developers (useful for debugging and recreating the object).

20. What is the significance of the super() function in Python?

super() allows a child class to call a method from the parent class. It is most commonly used in __init__ to ensure the parent class is set up correctly before the child adds its own data.

21. What is the significance of the __del__ method in Python?

The __del__ method is the destructor. It runs automatically when an object is about to be destroyed (garbage collected) to perform cleanup tasks like closing files or releasing resources.

22. What is the difference between @staticmethod and @classmethod?

@classmethod can access and modify class state (using cls). @staticmethod cannot access class or instance data; it acts as an independent utility function that lives inside the class namespace.

23. How does polymorphism work in Python with inheritance?

When a child class overrides a parent method, Python automatically calls the child's version at runtime. This allows you to loop through a list of different objects (parents and children) and call the same method name, getting the correct unique behavior for each.

24. What is method chaining in Python OOP?

Method chaining is calling multiple methods sequentially (e.g., obj.move().turn()). It is achieved by making each method return self at the end, passing the object context to the next call.

25. What is the purpose of the __call__ method in Python?

The __call__ method allows an object instance to be treated like a function. If you define this method, you can execute code by simply writing object_name() with parentheses.

```python
# 2.Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
# from it and implement the area() method in both.
from abc import ABC, abstractmethod
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass
class Circle(Shape):
  def area(self,r):
    return 3.14*r*r
class Rectangle(Shape):
  def area(self,l,b):
    return l*b
c = Circle()
print(c.area(5))
r = Rectangle()
print(r.area(5,6))
```

```python
# 3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
# and further derive a class ElectricCar that adds a battery attribute.
class Vehicle:
  def __init__(self,type):
    self.type = type
  def display(self):
    print(self.type)
class Car(Vehicle):
  def __init__(self,name,type):
    self.name = name
    super().__init__(type)
  def display(self):
    print(f"this car name is {self.name} and it is {self.type}")
class ElectricCar(Car):
  def __init__(self, battery, name, type):
    self.battery = battery
    super().__init__(name,type)
  def display(self):
    print(f"this car named {self.name} battery is {self.battery} bcz it is {self.type}")
v = Vehicle('Car')
c = Car('Innova','Toyota')
e = ElectricCar('5000mAh', 'Tesla', 'Electric Vehicle')
v.display()
c.display()
e.display()
```

```python
# 4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
# Sparrow and Penguin that override the fly() method.
class Bird():
  def fly(self):
    print('Bird can fly')
class Sparrow(Bird):
  def fly(self):
    print('Sparrow can fly')
class Penguin(Bird):
  def fly(self):
    print('Penguin can not fly')
b = Bird()
s = Sparrow()
p = Penguin()
b.fly()
s.fly()
p.fly()
```

```python
# 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
# balance and methods to deposit, withdraw, and check balance.
class BankAccount():
  def __init__(self,balance):
    self.__balance = 10000
  def deposite(self,amt):
    self.__balance += amt
  def withdraw(self,amt):
    if amt > self.__balance:
      print("Insufficient balance")
    else:
      self.__balance -= amt
  def check_balance(self):
    return self.__balance
account = BankAccount(1000)
account.deposite(500)
account.withdraw(200)
print(account.check_balance())
```

```python
# 7.Create a class MathOperations with a class method add_numbers() to add two numbers and a static
# method subtract_numbers() to subtract two numbers.
class MathOperation():
  @classmethod
  def add(cls, a, b):
    return a+b
  @staticmethod
  def sub(a,b):
    return a-b
math = MathOperation()
print(math.add(5,6))
print(math.sub(5,6))
```

```python
# 12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute
# the average of the grades
class Student:
  def __init__(self,name,grades):
    self.name = name
    self.grades = grades
  def avg_grade(self):
    sum = 0
    n = len(self.grades)
    for i in self.grades:
      sum += i
    return sum/n
s = Student("John",[50,60,70,24])
print(s.avg_grade())
```

```python
# 14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
# and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class Employee:
  def __init__(self,name,hours_worked,hourly_rate):
    self.name = name
    self.hours_worked = hours_worked
    self.hourly_rate = hourly_rate
  def calculate_salary(self):
    return self.hours_worked*self.hourly_rate
class Manager(Employee):
  def __init__(self,name,hours_worked,hourly_rate,bonus):
    super().__init__(name,hours_worked,hourly_rate)
    self.bonus = bonus
  def calculate_salary(self):
    return super().calculate_salary() + self.bonus
e = Employee("John",40,50)
print(e.calculate_salary())
m = Manager("John",40,50,1000)
print(m.calculate_salary())
```

```python
# 15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that
# calculates the total price of the product.
class Product:
  def __init__(self,name,price,quantity):
    self.name = name
    self.price = price
    self.quantity = quantity
  def total_price(self):
    return self.price*self.quantity
p = Product("Apple",10,5)
print(p.total_price())
```

```python
# 17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that
# returns a formatted string with the book's details.
class Book:
  def __init__(self,title,author,year_published):
    self.title = title
    self.author = author
    self.year_published = year_published
  def get_book_info(self):
    return f"Title: {self.title}\nAuthor: {self.author}\nYear Published: {self.year_published}"

b = Book("The Alchemist","Paulo Coelho",1988)
print(b.get_book_info())
```

```python
# 18. Create a class House with attributes address and price. Create a derived class Mansion that adds an
# attribute number_of_rooms.
class House:
  def __init__(self,address,price):
    self.address = address
    self.price = price
  def get_house_info(self):
    return f"Address: {self.address}\nPrice: {self.price}"
class Mansion(House):
  def __init__(self,address,price,number_of_rooms):
    super().__init__(address,price)
    self.number_of_rooms = number_of_rooms
  def get_mansion_info(self):
    return f"{super().get_house_info()}\nNumber of rooms: {self.number_of_rooms}"
h = House("123 Main St",100000)
print(h.get_house_info())
m = Mansion("123 Main St",100000,5)
print(m.get_mansion_info())
```
