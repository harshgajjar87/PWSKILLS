1. Difference between interpreted and compiled languages

Compiled languages convert the entire source code into machine code before execution (e.g., C, C++), making them faster.
Interpreted languages execute code line by line at runtime (e.g., Python), which makes debugging easier but execution slower.

2. What is exception handling in Python

Exception handling is a mechanism to handle runtime errors using try, except, else, and finally.
It prevents program crashes and allows graceful error handling.

3. Purpose of the finally block

The finally block always executes whether an exception occurs or not.
It is mainly used for cleanup operations like closing files or releasing resources.

4. What is logging in Python

Logging is used to record program events such as errors, warnings, and information messages.
It helps in debugging and monitoring applications without interrupting execution.

5. Significance of the __del__ method

The __del__ method is a destructor that runs when an object is deleted.
It is used to free resources like memory or close files.

6. Difference between import and from … import

import module imports the entire module and requires module.name to access members.
from module import name imports specific items and allows direct usage.

7. Handling multiple exceptions in Python

Multiple exceptions can be handled using multiple except blocks or a tuple of exceptions.
This helps manage different error types separately.

8. Purpose of the with statement in file handling

The with statement ensures proper resource management.
It automatically closes the file after the block executes, even if an error occurs.

9. Difference between multithreading and multiprocessing

Multithreading uses multiple threads within one process and shares memory.
Multiprocessing uses multiple processes with separate memory, suitable for CPU-bound tasks.

10. Advantages of using logging

Logging helps track errors, monitor application behavior, and debug efficiently.
It provides persistent records without affecting program flow.

11. What is memory management in Python

Memory management refers to allocation and deallocation of memory for objects.
Python handles this automatically using reference counting and garbage collection.

12. Basic steps in exception handling

Place risky code in try

Handle errors using except

Optional else runs if no error

finally executes cleanup code

13. Why memory management is important

Proper memory management prevents memory leaks and crashes.
It ensures efficient use of system resources and program stability.

14. Role of try and except

try contains code that may cause an error.
except handles the error gracefully without stopping the program.

15. How Python’s garbage collection works

Python uses reference counting to track object usage.
When references drop to zero, memory is freed, and cyclic garbage is handled by GC.

16. Purpose of the else block in exception handling

The else block runs only if no exception occurs in try.
It is used for code that should execute after successful execution.

17. Common logging levels in Python

Common levels are DEBUG, INFO, WARNING, ERROR, and CRITICAL.
They represent the severity of events.

18. Difference between os.fork() and multiprocessing

os.fork() creates a child process in Unix systems only.
The multiprocessing module is cross-platform and safer to use.

19. Importance of closing a file

Closing a file releases system resources.
It prevents data corruption and memory leaks.

20. Difference between file.read() and file.readline()

file.read() reads the entire file content at once.
file.readline() reads one line at a time.

21. Purpose of the logging module

The logging module provides a flexible way to log messages.
It supports different levels, formats, and output destinations.

22. Use of the os module in file handling

The os module allows file and directory operations.
Examples include file deletion, renaming, and directory management.

23. Challenges in memory management in Python

Circular references and large objects can cause memory issues.
Improper handling may lead to increased memory usage.

24. Raising an exception manually

Use the raise keyword to trigger an exception.
Example: raise ValueError("Invalid input")

25. Importance of multithreading

Multithreading improves application responsiveness.
It allows parallel execution of tasks, especially in I/O-bound applications.

```python
#7
import logging
logging.basicConfig(filename= "transaction_log.log", level = logging.DEBUG, force=True)

class BankAccount:
  def __init__(self,name,balance):
    self.name = name
    self.balance = balance
    logging.info(f"Account created for {name} with balance {balance}")
  def deposit(self,amount):
    self.balance += amount
    logging.info(f"{self.name} deposited {amount}. New balance is {self.balance}")
  def withdraw(self,amt):
    if amt > self.balance:
      logging.error(f"Withdrawl failed for {self.name}. Insufficient funds. Requested: {amt}, Available: {self.balance}")
    else:
      self.balance = self.balance - amt
      if self.balance < 500:
        logging.warning(f'withdraw successful. {self.name} has low balance: {self.balance}')
      else:
        logging.info(f"{self.name} withdrew {amt}. new balance: {self.balance}")
my_acc = BankAccount("Raj",20000)
my_acc.deposit(500)
my_acc.withdraw(2000)
my_acc.withdraw(18000)
my_acc.withdraw(300)
my_acc.withdraw(400)
```

```python
#14
import logging

# Configure logging to show the message level and the message
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

def process_data(data):
    logging.info("Started processing data...")

    if not data:
        logging.error("Data processing failed: No data provided!")
    else:
        logging.info(f"Data processed successfully: {data}")

# Test the logging
process_data("Sample Data")  # Logs INFO
process_data("")             # Logs INFO then ERROR
```
